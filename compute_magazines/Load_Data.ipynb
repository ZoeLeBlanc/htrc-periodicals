```python
def combine_pages(rows):
    final_row = rows[0:1].to_dict()
    tokens = ''
    for index, row in rows.iterrows():
        token_count = row['count']
        for i in range(0, int(token_count)):
            tokens += row.token + ' '
    final_row['positions'] = list(rows.pos)
    final_row['counts'] = list(rows['count'])
    final_row['sections'] = list(rows.section)
    final_row['tokens'] = tokens
    # print(pd.DataFrame([final_row]))
    # return tokens, positions, counts, sections
    return pd.DataFrame([final_row])
```
